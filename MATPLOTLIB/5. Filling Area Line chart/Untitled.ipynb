# plot, fill_between, alpha, median, where, interpolate,label

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('book.csv')
df.head(2)
```

```python
ages=df['Age']
python=df['Python']
js=df['JavaScript']
All=df['All_Devs']
```

```python
labels=['age','python']
labels2=['age','all']
plt.plot(ages,python,label=labels,linewidth=3)

plt.plot(ages,All,label=labels2,color='red',linestyle='--')
plt.fill_between(ages,python)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
labels=['age','python']
labels2=['age','all']
plt.plot(ages,python,label=labels,linewidth=3)

plt.plot(ages,All,label=labels2,color='red',linestyle='--')
plt.fill_between(ages,python,alpha=0.25)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
labels=['age','python']
labels2=['age','all']
plt.plot(ages,python,label=labels,linewidth=3)

#df['All_Devs'].median()
median=77616.0
plt.plot(ages,All,label=labels2,color='red',linestyle='--')
plt.fill_between(ages,python,median,alpha=0.25)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
labels=['age','python']
labels2=['age','all']
plt.plot(ages,python,label=labels,linewidth=3)

#df['All_Devs'].median()
median=77616.0
plt.plot(ages,All,label=labels2,color='red',linestyle='--')
plt.fill_between(ages,python,median,alpha=0.25,where=python>median)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
labels=['age','python']
labels2=['age','all']
plt.plot(ages,python,label=labels,linewidth=3)

#df['All_Devs'].median()
median=77616.0
plt.plot(ages,All,label=labels2,color='red',linestyle='--')
plt.fill_between(ages,python,median,alpha=0.25,where=python>median)
plt.fill_between(ages,python,median,alpha=0.25,where=python<=median)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
labels=['age','python']
labels2=['age','all']
plt.plot(ages,python,label=labels,linewidth=3)

#df['All_Devs'].median()
#median=77616.0
plt.plot(ages,All,label=labels2,color='red',linestyle='--')
plt.fill_between(ages,python,All,alpha=0.25,where=python>All,label='above_avg')
plt.fill_between(ages,python,All,alpha=0.25,where=python<=All,label='below_avg')
plt.legend()
plt.tight_layout()
plt.show()
```
