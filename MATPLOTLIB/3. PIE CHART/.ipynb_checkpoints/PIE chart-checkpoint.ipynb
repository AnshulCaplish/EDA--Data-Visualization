## Pie, labels,colors,wedgeprops,edgecolor,explode,startangle,autopct,shadow

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")
```

```python
slice=[60,40]
# pie plot will plot the percent of the individual value not the value. if the value is [120,80] plot will be same.
plt.pie(slice)
plt.title("pie chart")
plt.tight_layout()
plt.show()
```

```python
slice=[120,80]
labels=['Sixty %','Forty %']
plt.pie(slice,labels=labels)
plt.title("pie chart")
plt.tight_layout()
plt.show()
```

```python
# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
slice=[120,80,45,99]
labels=['Sixty %','Forty %','forty five','nine nine']
colors=['#008fd5','#fc4f30','#e5ae37','#6d904f']
# explode to mle a slice out of chart
explode = [0, 0, 0.1, 0]
# edgecolor=border color
plt.pie(slice,labels=labels,colors=colors,explode=explode,wedgeprops={'edgecolor':'black'})
plt.title("pie chart")
plt.tight_layout()
plt.show()
```

```python
slice=[120,80,45,99]
labels=['Sixty %','Forty %','forty five','nine nine']
colors=['#008fd5','#fc4f30','#e5ae37','#6d904f']
explode = [0, 0, 0.1, 0]
# startangle will rotate the chart
plt.pie(slice,labels=labels,colors=colors,explode=explode,startangle=180,wedgeprops={'edgecolor':'black'})
plt.title("pie chart")
plt.tight_layout()
plt.show()
```

```python
slice=[120,80,45,99]
labels=['Sixty %','Forty %','forty five','nine nine']
colors=['#008fd5','#fc4f30','#e5ae37','#6d904f']
explode = [0, 0, 0.1, 0]
# startangle will rotate the chart
# shadow for shdow
# auto_pct for percent data
plt.pie(slice,labels=labels,colors=colors,explode=explode,startangle=180,autopct='%1.1f%%',
        shadow=True,wedgeprops={'edgecolor':'black'})
plt.title("pie chart")
plt.tight_layout()
plt.show()
```
